```python
import numpy as np

np.random.seed(19680801)

N = 100

x = np.random.rand(N)
y = np.random.rand(N)

# print x
# print y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.axes().set_aspect('equal')
# REFERENCE: https://matplotlib.org/examples/pylab_examples/equal_aspect_ratio.html
```

```python
plt.show()
```

```python
N = 200; s = 0.05
x = np.random.normal(1,s,N) 
y = np.random.normal(1,s,N) 
plt.scatter(x, y) 
plt.axes().set_aspect('equal')
plt.show()
```
